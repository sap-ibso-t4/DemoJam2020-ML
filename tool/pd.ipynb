```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('../data/raw/engine.csv')
```

```python
print(df)
```

```python
dfPre = df.select_dtypes(include=['float64', 'int64'])
```

```python
dfPre
```

```python
dfPre = dfPre.drop(columns=['material id', 'type', 'category', 'hierarchy category', 'level'])
```

```python
print(dfPre)
```

```python
print(dfPre.columns)
```

```python
df_norm=dfPre.apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x)))
```

```python
print(df_norm)
```

```python
df_norm.head
```

```python
df_norm.head()
```

```python
print(df_norm.head())
```

```python
df_head = list(df_norm.columns)
```

```python
df_head
```

```python
df[df_head] = df_norm
```
